```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("C:\\Users\\LENOVO\\OneDrive\\Desktop\\data_jobs.csv")
df.job_posted_date = pd.to_datetime(df.job_posted_date)
import ast

df = df[df["job_skills"].notna()]  # Remove NaNs to avoid errors

df["job_skills"] = df["job_skills"].apply(ast.literal_eval)
```

```python
df[(df["job_country"] == "Germany") & (df["job_title_short"] == "Data Analyst")]
df_da_ger = df[(df["job_country"] == "Germany") & (df["job_title_short"] == "Data Analyst")]
```

```python
df_da_ger["job_location"].value_counts().head(10).to_frame()





```

```python
df_graph = df_da_ger["job_location"].value_counts().head(10).to_frame()
sns.barplot(data=df_graph, x="count", y="job_location", hue="count", palette="dark:b_r", legend=False)
plt.title("Data Analyst job in Germany")
plt.ylabel("")
plt.xlabel("job posted")
```

```python
dict_column = {
    "job_work_from_home": "work from home",
    "job_no_degree_mention": "Degree Requirement",
    "job_health_insurance": "health insurance offered"
}

# Set up subplots
fig, ax = plt.subplots(1, len(dict_column), figsize=(15, 5))

for i, (column, title) in enumerate(dict_column.items()):
    values = df_da_ger[column].value_counts()
    labels = values.index

    ax[i].pie(values, startangle=90, autopct="%1.1f%%", labels=labels)
    ax[i].set_title(title)

plt.tight_layout()
plt.show()

   
```

```python
df_graph = df_da_ger["company_name"].value_counts().head(10).to_frame()
sns.barplot(data=df_graph, x="count", y="company_name", hue="count", palette="dark:b_r", legend=False)
plt.title("counts of company")
plt.ylabel("")
plt.xlabel("job posted")
```
